```python
def simplifier(code):
    simplified = [['S','S'],['Sa','S'],['C','C'],['Ca','Ca'],['P','P'],['Pa','Pa']]
    codes = code.split('.')
    new_code = code
    for c in codes:
        for pair in simplified:
            if c == pair[0]:
                new_code = pair[1]
                break
        if new_code != code:
            break

    return new_code
```

```python
#Search returns statistics for syntactic structures inside of the cola

total_ct = 0
inventory = collections.OrderedDict([])

with open ('BHQchapters_1-8.csv','r') as csvfile:
    reader = csv.reader(csvfile)
    for line in reader:
        total_ct += 1
        code = simplifier(line[5])
        if code not in inventory:
            inventory[code] = 1
        elif code in inventory:
            inventory[code] += 1
            
inventory2 = collections.OrderedDict(sorted(inventory.items(), key=lambda t: t[1], reverse = True))

with open('BHQ_results.csv','w')as csvfile:
    writer = csv.writer(csvfile)
    
    for key in inventory2:
        percent = round((inventory2[key] / total_ct)*100,2)
        
        writer.writerow([key]+[inventory[key]]+[str(percent)+'%'])
        
for key in inventory2:
    percent = round((inventory2[key] / total_ct)*100,2)
    print ('{:14}{:14}{:14}%'.format(key, inventory[key], percent))
    
# Prints the less simple results

print ('\n')

total_ct = 0
inventory = collections.OrderedDict([])
    
with open ('BHQchapters_1-8.csv','r') as csvfile:
    reader = csv.reader(csvfile)
    for line in reader:
        total_ct += 1
        code = line[5]
        if code not in inventory:
            inventory[code] = 1
        elif code in inventory:
            inventory[code] += 1
            
inventory2 = collections.OrderedDict(sorted(inventory.items(), key=lambda t: t[1], reverse = True))
            
for key in inventory2:
    percent = round((inventory2[key] / total_ct)*100,2)
    print ('{:14}{:14}{:14}%'.format(key, inventory[key], percent))
```

# Cola with No Syntactic Correspondence

```python
# goal of the searches:
# identify syntactic structures held within 'none' cola boundaries
# store the node data for clause analysis
# return statistics for what kinds of structures are being labeled 'none'
```

```python
#indexes all nodes in Song of Songs for faster searches

corpus = 'Canticum'
cur_book = None
nodes = []

for n in NN():
    
    if cur_book == corpus:
        nodes.append(n)
    
    if F.otype.v(n) == 'book':
        cur_book = F.book.v(n)
        
msg('{} nodes appended'.format(len(nodes)))
```

```python
#indexes all monad sets and the monads in between that carry the "N" tag in the source file

filename = 'BHQchapters_1-8.csv'

monads = []
with open (filename, 'r') as csvfile:
    reader = csv.reader(csvfile)
    
    for line in reader:
        if line[5] == "N":
            monads.append([ x for x in range( int(line[1] ),int(line[2]) +1) ])
            
msg('{} monad sets added'.format(len(monads)))
```

```python
#cycles through every (sequential) syntactic possibiltiy within a monad set and returns the otypes found

results = []
sysct = 0
for lst in monads:
    result = []
    total_ct = 0
    for item in lst:    
        list_ct = len(lst) - total_ct
        ct = 0
        while True:
            if list_ct == 0:
                break
            elif ct != 0:
                monad_s = str(item)+"-"+str(item + ct)
            elif ct == 0:
                monad_s = str(item)
            for node in nodes:
                if F.monads.v(node) == monad_s:
                    result.append([monad_s, F.otype.v(node)])
            list_ct -= 1
            ct += 1
        total_ct += 1
    results.append(result)
    sysct += 1
msg("Complete")
```

```python
#Consolidates the repeated items into one entry

new_results = collections.OrderedDict([])
for monads in results:

    for lst in monads:
        
        if lst[0] not in new_results:
            new_results[lst[0]] = [lst[1]] 
        
        elif lst[0] in new_results:
            new_results[lst[0]].append(lst[1])
msg ("Complete")
```

```python
#writes data to a csv file and condenses otype labels into dot separated codes
#entries in csv are monad (reference, otype codes)

type_codes = {"word":"W", "phrase_atom":"Pa", "clause_atom":"Ca", "sentence_atom":"Sa", "subphrase":"Sp", 
              "phrase":"P","clause":"C","sentence":"S","half_verse":"Hv", "verse": "V"}

with open('none_type_syntax.csv','w') as csvfile:
    writer = csv.writer(csvfile)
    
    for key in new_results:
        abbreviation = ""
        dotct = len(new_results[key])
        
        for value in new_results[key]:
            
            dotct -= 1
            
            if dotct > 0:
                abbreviation += type_codes[value]+"."
            else:
                abbreviation += type_codes[value]
            
        writer.writerow([key]+[abbreviation])
```

```python
#retrieves (again) all monad references carrying the 'N' tag 
#(i.e. where no syntactic structures are found)

keys = []
with open(filename, 'r') as csvfile:
    reader = csv.reader(csvfile)
    
    for line in reader:
        if line[5] == "N":
            keys.append([line[1],line[2]])
```

```python
#maps monad references carrying 'N' tag as dict keys carrying empty values

none_types = collections.OrderedDict([])

for mon_lst in keys:
    
    none_types["{}-{}".format(mon_lst[0],mon_lst[1])] = []
    
msg("complete!")
```

```python
#reads the previously written csv file containing the condensed syntactic codes and their 
#monad references.
#maps the syntactic structures to the monad sets containing 'N' tag

for key in none_types:
    monads = key.split("-")
    monad_list = [str(x) for x in range(int(monads[0]), int(monads[1]) +1 )]
    
    with open ('none_type_syntax.csv', 'r') as csvfile:
        reader = csv.reader(csvfile)
    
        for line in reader:
            if line[0] == 6:
                if int(line[0]) in monad_list:
                    none_types[key].append([line[0],line[1]])

            else:
                split = line[0].split("-")
                if split[0] in monad_list:
                    none_types[key].append([line[0],line[1]])
msg('Complete')
```

```python
#the data is finished
#now we write all of the data to csv file with two sets
#data can then be reloaded as a dict with a key and value (see next cell)

with open ('none_type_syntax_coded.csv', 'w') as csvfile:
    writer = csv.writer(csvfile)
    
    for key in none_types:
        writer.writerow([key] + [none_types[key]])
```

```python
#example data loader

import ast #enables list to be imported from csv as dict value

none_dict = collections.OrderedDict([])

with open('none_type_syntax_coded.csv', 'r') as csvfile:
    reader = csv.reader(csvfile)
    
    for line in reader:
        none_dict[line[0]] = ast.literal_eval(line[1]) #ast.literal_eval interprets str as code
    
    msg('done')
```

```python
#now analyzing data set
#what types of structures are passed over resulting in the 'none' label?

counts = collections.OrderedDict([])
none = 0
none_list = []
curious = []

for key in none_dict:
    sentence_ct = 0
    clause_ct = 0
    
    for pair in none_dict[key]:
        search = pair[1].split('.')

        if "S" in search:
            sentence_ct += 1
        elif "C" in search:
            clause_ct += 1
                         
    if sentence_ct == 0 and clause_ct == 0:
        none += 1
        none_list.append(key)
    
    else:
        if ('S:{}'.format(sentence_ct) in counts) and (sentence_ct != 0):
            counts['S:{}'.format(sentence_ct)] += 1
            
        elif ('C:{}'.format(clause_ct) in counts) and clause_ct != 0:
            counts['C:{}'.format(clause_ct)] += 1
            
        elif ('S:{}'.format(sentence_ct) not in counts) and (sentence_ct != 0):
            counts['S:{}'.format(sentence_ct)] = 1
            
        elif ('C:{}'.format(clause_ct) not in counts) and clause_ct != 0:
            counts['C:{}'.format(clause_ct)] = 1
            
        if sentence_ct == 2:
            curious.append(key)
            
msg('finished!')

for pair in curious:
    pair = pair.split('-')
    x = int(pair[0])
    print ('{}:{:10}{:10}\n'.format(F.chapter.v(L.u("chapter", x) ),
                                   F.verse.v( L.u("verse", x)),
                                   T.words(L.d('word', (L.u("sentence",x))),fmt = 'ha')
                                  ))
```

```python
#prints the results of the analysis in a clean, hierarchical format

print ("cola with no syntactic correspondence ('none'): {:16}  21.1% of all cola\n".format(len(none_dict)))

print ("      cola as 'none', containing <1 syntactic unit: {:12}  {}% of 'none'\n".format(
        len(none_list),
        str(round(float(len(none_list) / len(none_dict))*100,2)))
      )

print ("      cola as 'none', containing >1 syntactic unit: {:12}  {}% of 'none'".format(sum([counts[key] for key in counts]),
                                                                   round(float(sum([counts[key] for key in counts]) / len(none_dict))*100,2)
                                                         ))

counts = collections.OrderedDict(sorted(counts.items(), key=lambda t: t[0]))


for key in counts:
    splitter = key.split(':')
    if splitter[0] == "S":
        print ("           {:5} cola with {} sentence(s)".format(counts[key],splitter[1]))
    else:
        print ("           {:5} cola with {} clause(s)".format(counts[key],splitter[1]))
```
